```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("./kidney_disease.csv")
dataset.head()
```

```python
dataset = dataset.drop_duplicates()
```

```python
dataset = dataset.dropna(subset = ['rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane'])
```

```python
dataset.isnull().any()
```

```python
dataset['age'] = dataset['age'].fillna(int(dataset['age'].mean()))
dataset['bp'] = dataset['bp'].fillna(int(dataset['bp'].mean()))
dataset['sg'] = dataset['sg'].fillna(dataset['sg'].mean())
dataset['al'] = dataset['al'].fillna(int(dataset['al'].mean()))
dataset['su'] = dataset['su'].fillna(int(dataset['su'].mean()))
dataset['bgr'] = dataset['bgr'].fillna(int(dataset['bgr'].mean()))
dataset['bu'] = dataset['bu'].fillna(int(dataset['bu'].mean()))
dataset['sc'] = dataset['sc'].fillna((dataset['sc'].mean()))
dataset['sod'] = dataset['sod'].fillna(int(dataset['sod'].mean()))
dataset['pot'] = dataset['pot'].fillna((dataset['pot'].mean()))
dataset['hemo'] = dataset['hemo'].fillna((dataset['hemo'].mean()))
dataset['pcv'] = dataset['pcv'].fillna(int(dataset['pcv'].mean()))
dataset['wc'] = dataset['wc'].fillna(int(dataset['wc'].mean()))
dataset['rc'] = dataset['rc'].fillna(int(dataset['rc'].mean()))
```

```python
dataset.isnull().any()
```

```python
# 'rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane','classification'
print(dataset['rbc'].unique())
print(dataset['pc'].unique())
print(dataset['pcc'].unique())
print(dataset['ba'].unique())
print(dataset['htn'].unique())
print(dataset['dm'].unique())
print(dataset['cad'].unique())
print(dataset['appet'].unique())
print(dataset['pe'].unique())
print(dataset['ane'].unique())
print(dataset['classification'].unique())
```

```python
dataset['rbc'] = dataset['rbc'].replace('normal',0)
dataset['rbc'] = dataset['rbc'].replace('abnormal',1)

dataset['pc'] = dataset['pc'].replace('normal',0)
dataset['pc'] = dataset['pc'].replace('abnormal',1)

dataset['pcc'] = dataset['pcc'].replace('notpresent',0)
dataset['pcc'] = dataset['pcc'].replace('present',1)

dataset['ba'] = dataset['ba'].replace('notpresent',0)
dataset['ba'] = dataset['ba'].replace('present',1)

dataset['htn'] = dataset['htn'].replace('no',0)
dataset['htn'] = dataset['htn'].replace('yes',1)

dataset['dm'] = dataset['dm'].replace('no',0)
dataset['dm'] = dataset['dm'].replace('yes',1)

dataset['cad'] = dataset['cad'].replace('no',0)
dataset['cad'] = dataset['cad'].replace('\tno',0)
dataset['cad'] = dataset['cad'].replace('yes',1)

dataset['appet'] = dataset['appet'].replace('good',0)
dataset['appet'] = dataset['appet'].replace('poor',1)

dataset['pe'] = dataset['pe'].replace('no',0)
dataset['pe'] = dataset['pe'].replace('yes',1)

dataset['ane'] = dataset['ane'].replace('no',0)
dataset['ane'] = dataset['ane'].replace('yes',1)

dataset['classification'] = dataset['classification'].replace('notckd',0)
dataset['classification'] = dataset['classification'].replace('ckd',1)
dataset['classification'] = dataset['classification'].replace('ckd\t',1)
```

```python
print(dataset['rbc'].unique())
print(dataset['pc'].unique())
print(dataset['pcc'].unique())
print(dataset['ba'].unique())
print(dataset['htn'].unique())
print(dataset['dm'].unique())
print(dataset['cad'].unique())
print(dataset['appet'].unique())
print(dataset['pe'].unique())
print(dataset['ane'].unique())
print(dataset['classification'].unique())
```
